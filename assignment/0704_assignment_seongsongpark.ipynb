```python
def changetonum(x):
    if x == 'A+': 
        return 4.5
    elif x == 'A':
        return 4.0
    elif x =='B+':
        return 3.5
    elif x =='B':
        return 3.0
    elif x == 'C+':
        return 2.5
    elif x == 'C':
        return 2.0
    elif x == 'D+':
        return 1.5
    elif x =='D':
        return 1.0
    elif x =='F':
        return 0
    else :
        printf("성적을 잘 못 입력하였습니다")

sum_grade = 0.0
count = 0.0
while True:
    grade = input("성적을 입력하세요")
    if grade :
        if grade.isupper() :
            count+=1
            sum_grade+=changetonum(grade)
            
        else:
            print("대문자로 올바르게 입력하세요")
    else :
        print(sum_grade/count)
        break;
    
    
```

```python
# Exercise 67

def decide_price(x) :
    if x<3 :
        return 0
    elif 2<x<13:
        return 14
    elif x>=65 :
        return 18
    else :
        return 23

sum_price=0
while True:
    age_s = input(" 나이를 입력하세요")
    if age_s :
        try :
            age = int(age_s)
        except :
            print("숫자를 제대로 입력하세요")
            
        sum_price += decide_price(age)
    else :
        print(sum_price)
        break;
```
